```python
class MSE(nn.Module):
  def __init__(self,y_k,n):
    super().__init__()
    self.y_k = y_k
    self.n = n
  def forward(self,y):
    z = []
    z2 = []
    for i,m in self.y_k,y:
      m1 = i - m
      z.append(m1)
    for u in z:
      m2 = u**2
      z2.append(m2)
    m3 = sum(z2)
    m4 = m3/self.n
    return m4
```

```python
def one_hot_encode(data):
    """
    Преобразует список категориальных значений в one-hot формат.

    Параметр:
    - data: список значений, например ['красный', 'синий', 'зелёный', 'красный']

    Возвращает:
    - encoded: список списков (матрица), где каждый подсписок — one-hot вектор
    - categories: список уникальных категорий (порядок важен!)
    """

    # Шаг 1: Найти уникальные категории в том порядке, в котором они встречаются
    categories = []
    for item in data:
        if item not in categories:
            categories.append(item)

    # Шаг 2: Создать словарь: категория -> индекс
    category_to_index = {}
    for idx, category in enumerate(categories):
        category_to_index[category] = idx

    # Шаг 3: Преобразовать каждое значение в one-hot вектор
    encoded = []
    n_categories = len(categories)

    for item in data:
        # Создаём вектор из нулей
        one_hot = [0] * n_categories
        # Ставим 1 в позиции соответствующей категории
        one_hot[category_to_index[item]] = 1
        # Добавляем в результат
        encoded.append(one_hot)

    return encoded, categories
```

```python
import math

def categorical_cross_entropy(y_true, y_pred, epsilon=1e-15):
    """
    Вычисляет категориальную кросс-энтропию для батча.

    Параметры:
    - y_true: список списков, one-hot метки (например, [[0,1,0], [1,0,0]])
    - y_pred: список списков, предсказанные вероятности (должны суммироваться в 1 по строке)
    - epsilon: малое число, чтобы избежать log(0)

    Возвращает:
    - Среднее значение потерь по всем примерам
    """
    n_samples = len(y_true)
    total_loss = 0.0

    for i in range(n_samples):
        sample_loss = 0.0
        for j in range(len(y_true[i])):
            # "Подрезаем" предсказания, чтобы избежать log(0)
            pred = y_pred[i][j]
            if pred < epsilon:
                pred = epsilon
            elif pred > 1 - epsilon:
                pred = 1 - epsilon

            # Вклад j-го класса в loss: y_true * log(pred)
            sample_loss += y_true[i][j] * math.log(pred)

        # Кросс-энтропия — это минус сумма
        total_loss -= sample_loss

    # Среднее по всем образцам
    return total_loss / n_samples
```
