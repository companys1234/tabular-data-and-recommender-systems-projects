```python
x = [1,2,3]
import math
import numpy as np
```

```python
def batchnorm(b):
  d = np.random.rand()
  bt = np.random.rand()
  e = 10**(-8)
  u = sum(b)*(1/len(b))
  z = 0
  for i in b:
    z += (i-u)**2
  q = z*(1/len(b))
  l = []
  for i in b:
    p = (i-u)/(math.sqrt(q+e))
    l.append(p)
  kl = []
  for i in l:
    y = i*d + bt
    kl.append(y)
  return kl
```

```python
batchnorm(x)
```
