```python
def load_market_basket_data():
    """Загрузка примера данных рыночной корзины"""
    # Пример данных (товары в супермаркете)
    transactions = [
        ['молоко', 'хлеб', 'печенье'],
        ['молоко', 'хлеб', 'масло', 'яйца'],
        ['хлеб', 'масло', 'печенье'],
        ['молоко', 'хлеб', 'масло', 'печенье'],
        ['хлеб', 'печенье'],
        ['молоко', 'хлеб'],
        ['молоко', 'масло'],
        ['хлеб', 'масло'],
        ['молоко', 'хлеб', 'масло', 'яйца', 'печенье'],
        ['молоко', 'хлеб', 'масло']
    ]
    return transactions


def load_online_retail_data():
    """Загрузка данных онлайн-ритейла (упрощенный пример)"""
    # Генерируем синтетические данные
    np.random.seed(42)

    products = [
        'Ноутбук', 'Мышь', 'Клавиатура', 'Наушники', 'Монитор',
        'Кабель USB', 'Коврик для мыши', 'Веб-камера', 'Микрофон',
        'Колонки', 'Принтер', 'Сканер', 'Планшет', 'Смартфон'
    ]

    transactions = []
    for _ in range(100):
        n_items = np.random.randint(2, 6)
        transaction = np.random.choice(products, n_items, replace=False).tolist()
        transactions.append(transaction)

    return transactions


def demo_apriori_from_scratch():
    """Демонстрация Apriori с нуля"""
    print("=" * 60)
    print("ДЕМОНСТРАЦИЯ APRIORI (РЕАЛИЗАЦИЯ С НУЛЯ)")
    print("=" * 60)

    # Загружаем данные
    transactions = load_market_basket_data()

    # Инициализируем и обучаем модель
    model = Apriori(min_support=0.3, min_confidence=0.7)
    model.fit(transactions)

    # Получаем правила
    print("\nАССОЦИАТИВНЫЕ ПРАВИЛА:")
    print("-" * 60)
    rules_df = model.get_rules(10)
    print(rules_df.to_string(index=False))

    # Выводим статистику
    print("\nСТАТИСТИКА:")
    print("-" * 60)
    total_itemsets = sum(len(itemsets) for itemsets in model.frequent_itemsets.values())
    print(f"Всего транзакций: {model.n_transactions}")
    print(f"Найдено частых наборов: {total_itemsets}")
    print(f"Найдено правил: {len(model.rules)}")

    # Визуализация
    model.visualize_itemsets()

    return model


def demo_fpgrowth_from_scratch():
    """Демонстрация FP-Growth с нуля"""
    print("\n" + "=" * 60)
    print("ДЕМОНСТРАЦИЯ FP-GROWTH (РЕАЛИЗАЦИЯ С НУЛЯ)")
    print("=" * 60)

    transactions = load_market_basket_data()

    model = FPGrowth(min_support=0.3, min_confidence=0.7)
    model.fit(transactions)

    print("\nАССОЦИАТИВНЫЕ ПРАВИЛА:")
    print("-" * 60)
    rules_df = model.get_rules(10)
    print(rules_df.to_string(index=False))

    return model


def demo_mlxtend_comparison():
    """Сравнение Apriori и FP-Growth с использованием mlxtend"""
    print("\n" + "=" * 60)
    print("СРАВНЕНИЕ APRIORI И FP-GROWTH (MLXTEND)")
    print("=" * 60)

    # Загружаем более крупные данные
    transactions = load_online_retail_data()

    # Тестируем Apriori
    print("\n1. APRIORI АЛГОРИТМ:")
    print("-" * 40)

    apriori_model = AssociationRulesMLxtend(
        algorithm='apriori',
        min_support=0.05,
        min_confidence=0.5
    )
    apriori_model.fit(transactions)

    print(f"Найдено частых наборов: {len(apriori_model.frequent_itemsets)}")
    print(f"Найдено правил: {len(apriori_model.rules)}")

    top_apriori = apriori_model.get_top_rules(5)
    print("\nТоп-5 правил (Apriori):")
    for i, (_, rule) in enumerate(top_apriori.iterrows(), 1):
        print(f"{i}. {set(rule['antecedents'])} → {set(rule['consequents'])}")
        print(f"   Confidence: {rule['confidence']:.3f}, Lift: {rule['lift']:.3f}")

    # Тестируем FP-Growth
    print("\n2. FP-GROWTH АЛГОРИТМ:")
    print("-" * 40)

    fpgrowth_model = AssociationRulesMLxtend(
        algorithm='fpgrowth',
        min_support=0.05,
        min_confidence=0.5
    )
    fpgrowth_model.fit(transactions)

    print(f"Найдено частых наборов: {len(fpgrowth_model.frequent_itemsets)}")
    print(f"Найдено правил: {len(fpgrowth_model.rules)}")

    top_fpgrowth = fpgrowth_model.get_top_rules(5)
    print("\nТоп-5 правил (FP-Growth):")
    for i, (_, rule) in enumerate(top_fpgrowth.iterrows(), 1):
        print(f"{i}. {set(rule['antecedents'])} → {set(rule['consequents'])}")
        print(f"   Confidence: {rule['confidence']:.3f}, Lift: {rule['lift']:.3f}")

    # Рекомендации
    print("\n3. СИСТЕМА РЕКОМЕНДАЦИЙ:")
    print("-" * 40)

    test_basket = ['Ноутбук', 'Мышь']
    recommendations = fpgrowth_model.recommend(test_basket, 3)

    print(f"Корзина: {test_basket}")
    print("\nРекомендации:")
    for i, rec in enumerate(recommendations, 1):
        print(f"{i}. {rec['item']} (доверие: {rec['confidence']:.3f}, lift: {rec['lift']:.3f})")

    # Визуализация
    print("\n4. ВИЗУАЛИЗАЦИЯ ПРАВИЛ:")
    fpgrowth_model.visualize_rules(15)

    return apriori_model, fpgrowth_model


def performance_comparison():
    """Сравнение производительности алгоритмов"""
    import time

    print("\n" + "=" * 60)
    print("СРАВНЕНИЕ ПРОИЗВОДИТЕЛЬНОСТИ")
    print("=" * 60)

    # Создаем большие данные для теста
    np.random.seed(42)
    products = [f"P{i:03d}" for i in range(100)]  # 100 товаров

    # Генерируем транзакции разного размера
    sizes = [100, 500, 1000]

    results = []

    for size in sizes:
        print(f"\nРазмер данных: {size} транзакций")
        print("-" * 40)

        # Генерируем транзакции
        transactions = []
        for _ in range(size):
            n_items = np.random.randint(2, 8)
            transaction = np.random.choice(products, n_items, replace=False).tolist()
            transactions.append(transaction)

        # Тестируем Apriori
        start_time = time.time()
        apriori_model = AssociationRulesMLxtend('apriori', 0.05, 0.5)
        apriori_model.fit(transactions)
        apriori_time = time.time() - start_time

        # Тестируем FP-Growth
        start_time = time.time()
        fpgrowth_model = AssociationRulesMLxtend('fpgrowth', 0.05, 0.5)
        fpgrowth_model.fit(transactions)
        fpgrowth_time = time.time() - start_time

        results.append({
            'size': size,
            'apriori_time': apriori_time,
            'fpgrowth_time': fpgrowth_time,
            'apriori_rules': len(apriori_model.rules) if apriori_model.rules is not None else 0,
            'fpgrowth_rules': len(fpgrowth_model.rules) if fpgrowth_model.rules is not None else 0
        })

        print(f"Apriori:   {apriori_time:.3f} сек, правил: {results[-1]['apriori_rules']}")
        print(f"FP-Growth: {fpgrowth_time:.3f} сек, правил: {results[-1]['fpgrowth_rules']}")

    # Визуализация результатов
    import matplotlib.pyplot as plt

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    sizes = [r['size'] for r in results]
    apriori_times = [r['apriori_time'] for r in results]
    fpgrowth_times = [r['fpgrowth_time'] for r in results]

    # График времени выполнения
    ax1.plot(sizes, apriori_times, 'o-', label='Apriori', linewidth=2)
    ax1.plot(sizes, fpgrowth_times, 's-', label='FP-Growth', linewidth=2)
    ax1.set_xlabel('Количество транзакций')
    ax1.set_ylabel('Время выполнения (сек)')
    ax1.set_title('Сравнение времени выполнения')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # График количества правил
    apriori_rules = [r['apriori_rules'] for r in results]
    fpgrowth_rules = [r['fpgrowth_rules'] for r in results]

    x = np.arange(len(sizes))
    width = 0.35

    ax2.bar(x - width/2, apriori_rules, width, label='Apriori', alpha=0.7)
    ax2.bar(x + width/2, fpgrowth_rules, width, label='FP-Growth', alpha=0.7)
    ax2.set_xlabel('Размер данных')
    ax2.set_ylabel('Количество правил')
    ax2.set_title('Количество найденных правил')
    ax2.set_xticks(x)
    ax2.set_xticklabels(sizes)
    ax2.legend()
    ax2.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

    return results


def main():
    """Главная функция демонстрации"""
    print("АССОЦИАТИВНЫЕ ПРАВИЛА: APRIORI И FP-GROWTH")
    print("=" * 60)

    # Демонстрация реализаций с нуля
    demo_apriori_from_scratch()
    demo_fpgrowth_from_scratch()

    # Демонстрация с использованием mlxtend
    demo_mlxtend_comparison()

    # Сравнение производительности
    performance_comparison()

    print("\n" + "=" * 60)
    print("ДЕМОНСТРАЦИЯ ЗАВЕРШЕНА!")
    print("=" * 60)


if __name__ == "__main__":
    main()
```
