```python
import numpy as np
import torch.nn as nn
```

```python
class lin_layer(nn.Module):
  def __init__(self,n,m):
    super().__init__()
    self.w = np.random.rand(m,n)
    self.b = np.random.rand(m,1)
  def forward(self,x):
    z = x @ self.w
    y = z + self.b
    return y
```

```python
p = np.random.rand(1,1)
print(p)
```

```python
l = lin_layer(1,1)
print(l(p))
```
